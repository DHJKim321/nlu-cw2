```python
with open('model_translations_baseline.txt', 'r') as f:
    baseline = f.read().split('\n')

with open('model_translations_lexical.txt', 'r') as f:
    lexical = f.read().split('\n')
```

```python
with open('output_comparison.txt', 'w') as f:
    for b, l in zip(baseline, lexical):
        f.write(f'Baseline: {b}\n')
        f.write(f'Lexical: {l}\n\n')
```
